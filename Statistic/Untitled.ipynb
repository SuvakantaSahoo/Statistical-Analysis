```python
import random

def median(sample):
    
    n=len(sample)
    if (n%2==0):
        return ((sample[n//2])+(sample[n//2-1]))/2
    else:
        return sample[n//2]

x=[]
for i in range(0,5):
    x.append(random.randint(10,50))
x.sort()
print('Values:',x)
print()
print('Median :',median(x))
```

### Outliers and IQR
- IQR= Q3-Q1
- Lower Bound = Q1 - (1.5*IQR)
- Upper Bound = Q3 + (1.5*IQR)

```python
q1=data['fare'].quantile(0.25)
q3=data['fare'].quantile(0.75)
iqr=q3-q1
print(q1)
print(q3)
print(iqr)
print()
lb=q1-(1.5*iqr)
ub=q3+(1.5*iqr)
print(lb,ub)
```

```python
len(data[(data['fare']<lb) | (data['fare']>ub)])
```

```python
data[(data['fare']<lb) | (data['fare']>ub)]
```

```python
def outlier(dataset,col):
    q1=dataset[col].quantile(0.25)
    q3=dataset[col].quantile(0.75)
    iqr=q3-q1
    lb=q1-(1.5*iqr)
    ub=q3+(1.5*iqr)
    outliers=list(dataset[(dataset[col]<lb) | (dataset[col]>ub)][col])
    outliers_len=len(outliers)
    print('Lenth of outliers of {} is: {}'.format(col,outliers_len))
    print('Outliers are :',outliers)

outlier(data,'fare')
```
